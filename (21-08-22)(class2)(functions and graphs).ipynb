```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,a,b,c):
    return  a*x**2+b*x+c
```

```python
xlist = np.linspace(-10,10,num=1000)
```

```python
a=5
b=1
c=4
ylist = f(xlist,a,b,c)
print(ylist)
```

```python
plt.figure(num=0,dpi=120)
plt.plot(xlist,ylist,label="f(x)")
plt.title("Plotting Example")
plt.xlabel("Distance/ft")
plt.ylabel("Height/ft")
plt.legend()
```

```python
def f1(x,a,b,c):
    return ((a/x)+(x-b)/(x**2 - c))
```

```python
xlist = np.linspace(-10,10,num=1000)
```

```python
a=3
b=1
c=1
ylist = f1(xlist,a,b,c)
print(ylist)
```

```python
plt.figure(num=0,dpi=120)
plt.plot(xlist,ylist,label="f(x)")
plt.title("Plotting Example")
plt.xlabel("Distance/ft")
plt.ylabel("Height/ft")
plt.legend()
```
